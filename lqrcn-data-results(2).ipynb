```python
import pandas as pd
import json
search_strings = ["民主", "民主主義", "民主化", "公民", "共和國", "國家", "憲法", "民權", "主權", "戰爭", "大戰","世界大戰", "戰", "失敗", "投降", "佔領", "侵占", "海軍", "陸軍", "空軍","日空軍", "日軍"]
```

```python
# Prepare a look up table which maps a link to a date

dates = pd.read_excel("/kaggle/input/lqrcn-data/lqrcn-data.xlsx")
dates = dates[["urls", "Date", "Newspaper"]]
dates = dates.set_index("urls")
dates_map = dates["Date"].to_dict()
list(dates_map.items())[:3]
```

```python
texts = json.load(open("/kaggle/input/lqrcn-data/run_results.json"))["links"]

all_results = []

for text_number, text in enumerate(texts):
    
    link = text["link"]
    # print(link)
    
    results = [link, dates_map[link]]
    
    # Check correctness of file format
    for key in text.keys():
        if not (key == "link" or key[:4] == "text"):
            raise ValueError("Object contains key other than link or text*")
        text.keys()
    
    # Connect all single texts to a complete text
    text_keys = [x for x in text.keys() if "text" in x]
    strings = [text[key] for key in text_keys]
    full_text = "\n\n".join(strings)
    
    for word in search_strings:
        # results.append(f'{word} {full_text.cound(word)}')
        results.append(full_text.count(word))
        
    all_results.append(results)
    
pd.DataFrame(all_results, columns=["Link", "Date"] + search_strings).to_excel("/kaggle/working/" + "test_output.xlsx")
```
